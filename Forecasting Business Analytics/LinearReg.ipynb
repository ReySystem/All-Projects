```python
#LINEAR REGRESSION 
```

```python
import pandas as pd 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/arib168/data/main/Salary_Data.csv')
df
```

```python
df.info()
```

```python
x = df.iloc[:,[0]].values # we require x to be in 2 dimensions
x  #.values converts data into a numpy array 
```

```python
x.ndim
```

```python
type(x)
```

```python
y = df.iloc[:,1].values
y
```

```python
y.ndim
```

```python
#we use library for ML called Sklearn/sci-kit learn 
from sklearn.linear_model import LinearRegression 
model = LinearRegression()    #PascalCase - First alphabet capital, the second word combined and first alphabet of the second word capital 
```

```python
model.fit(x,y)
```

```python
y_pred = model.predict(x)
y_pred   #predicted output 
```

```python
y   #real output 
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y) #input points
plt.plot(x,y_pred,c='r') #Best fit line 
plt.show()
```

```python
from sklearn.metrics import r2_score,mean_squared_error #regression score 
print(r2_score(y,y_pred))
print(mean_squared_error(y,y_pred))
```

```python
model.coef_     #slope (m) os the line is 9449.962
```

```python
model.intercept_    # c value is 23792.20
```
