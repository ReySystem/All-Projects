```python
#population - entire cholestrol column is the population (overall data)
#sampling - process of taking a few random data points from an entire column 
# lets take 10 samples from the chol column 
chol_series = df.chol
chol_series
print(chol_series.size)
```

```python
chol_sample = chol_series.sample(n=10)  #taking 10 random samples 
chol_sample
```

```python
#distribution plot for cholestrol samples
import matplotlib.pyplot as plt 
plt.figure(figsize=(8,5))
plt.hist(chol_series)
plt.xlabel("Cholestrol")
plt.ylabel("Frequency")
plt.title("Historgram of Cholestrol frequency")
plt.axvline(x = chol_series.mean(),color ='r')
print("Mean cholestrol value is :",chol_series.mean())
plt.show()
```

```python
#taking 300 samples from chol column 

mean_list = []
for i in range(300):
  
  mean_list.append(chol_series.sample(n = 30).mean())
print(len(mean_list))
print(mean_list)
```
